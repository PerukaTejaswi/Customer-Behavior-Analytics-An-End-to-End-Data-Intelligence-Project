```python
#loading csv data to pandas
import pandas as pd
df=pd.read_csv('customer_shopping_behavior.csv')
```

```python
#checking if any columns has null values
df.isnull().sum()
```

```python
#replacing null values with median of respective categories
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
#converting to snake case
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
#renaming a column which does not match snake_case
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#create new column age-group
labels=['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
df[['age','age_group']].head(10)
```

```python
#create a column purchase_frequency_days from frequency_of_purchases. string column to numeric values column
frequency_mapping={
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 Months':90
}
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(50)
```

```python
#checking if both the columns are the same - discount_applied and promo_code_used
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
df.columns
```

```python
#As they are the same, drop promo_code_used column
df=df.drop('promo_code_used', axis=1)
```
